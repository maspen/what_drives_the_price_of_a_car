```python
# Normalize data
mu = df.mean()
sigma = df.std()

df_norm = (df-mu)/sigma
df_norm
```

```python
# perform SVD on normalized data

U, s, Vt = svd(df_norm, full_matrices=False)

Sigma = np.diag(s)
V = Vt.T
```

```python
# verify that every entry in df_norm is close to the product of the matrices

np.allclose(df_norm, U @ Sigma @ V.T )
```
